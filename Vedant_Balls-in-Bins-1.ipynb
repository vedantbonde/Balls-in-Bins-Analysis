```
def ballsbins(n, b):
  # the function to list all the bins containing the total number of balls 
  # n: denotes the number of balls
  # b: denotes the number of distinct bins

    arr = np.identity(b, dtype=int)
    for c in itertools.combinations_with_replacement(arr, n): 
      # Combinations with replacements is the same as balls in distinct bins problem
        yield sum(c)

def ballsbins_list(n, b):
  # This function generates the list of all the configurations of bins in the form of an array
  output = np.array(list(ballsbins(n, b)))
  return output
```

```
def restrict(n, b, k):

  # Function which takes into account the restriction on the maximum amount of balls possible in a bin 
  # n: denotes the number of balls
  # b: denotes the number of distinct bins
  # k: denotes the restriction on the number of bins
  

  o = np.array(list(ballsbins(n, b)))
  it = o.shape[0]
  wid = o.shape[1]
  arr = np.arange(wid)
  arr = arr.reshape(1,wid)
  for i in range(it):
    if (max(o[i])==k):
      brr = o[i]
      brr = brr.reshape(1,wid)
      arr = np.concatenate((arr,brr), axis = 0)

  arr = np.delete(arr, 0, 0)

  return arr
```

```
def choose(n,k):

  # This is the function definitin of nCk or "n choose k" in combinatorics

  if (n<k or n<0):
    return 0
  elif (n == k and n==0):
    return 1
  else:
    result = int(fact(n)/(fact(n-k)*fact(k)))
    return result
```

```
def bins_restrict(n, b, r):
  
  # This function calculates the solution to x_1 + x_2 + ... + x_b = n, with the restriction that each bin contains atmost r elements 

  # n: Number of balls
  # b: Number of bins
  # r: Restriction on the bins 

  # This formula has been taken from the site: https://www.mathpages.com/home/kmath337/kmath337.htm , it also has a generalization for different capacities for different bins. 


  if (n==b and n==0):
    return 1
    # The above formula says that if there are 0 bins and 0 balls, then there is only 1 way to satisfy the constraint 
  
  elif (n<0):
    return 0

  else: 
    it = 0
    result = 0
    div = int((n/r)+1)

    for it in range(0,div+1):
      temp = ((-1)**it)*choose(b, it)*choose(n - it*(r+1) + b - 1, b - 1)
      # temp = ((-1)**it)*choose(b, it)*choose(n - it*(r+1) + b - 1, b - it*(r+1))
      result = result + temp

    return result
```

```
def B_ballsbins_calc(n, b, r):

  # This formula is useful for calculating the first case of Non-Dominant Balls and Bins problem. In this case, one of the bins has a dominant number of balls, i.e exactly r, 
  # and the rest have less than r balls, except for the case of b = 2. Note that this is only valid for the M_(2k,l,k), and possibly the dominant bins and balls 
  if (n < b):
    return 0
  else:
    temp = b
    new_balls = n - r
    new_bins = b - 1
    new_restrict = r - 1
    if (new_balls < new_bins):
      return 0
    else:
      new_balls = new_balls - (b - 1)
      cal = bins_restrict(new_balls, new_bins, new_restrict)
      temp = temp*cal
  #   if(b == 2):
  #     return int(temp/2)
  #   else:
      return temp
```

```
def M_2k_l_k(k, l):

# The definiton given in the paper for calculating number of ways to split 2k balls into l nonempty bins where 2 ≤ l ≤ k + 1 and the most crowded bin has exactly k balls

  if (2 <= l and l <= k+1):
    if (l == 2):
      return 1
    else:
      result = l*choose(k-1, l-2)
      return result
  else:
    return 0
```
